```python
#TASK: Create a box plot for the following data: data = [7, 8, 5, 6, 9, 10, 6, 7, 8, 9, 6].
#Customize the box plot by adding a title, labels for both axes, and a specific color for the box.
```

```python
import matplotlib.pyplot as plt

try:
    # Data
    data = [7, 8, 5, 6, 9, 10, 6, 7, 8, 9, 6]

    # Create the box plot
    plt.figure(figsize=(7, 5))
    box = plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor="lightblue"))

    # Customize the plot
    plt.title("Box Plot of Data")
    plt.xlabel("Data Set")
    plt.ylabel("Values")
    plt.grid(axis='y', linestyle="--", alpha=0.6)

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```

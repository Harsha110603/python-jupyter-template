### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for different plots
    x = np.arange(1, 6)
    y = x ** 2
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    scatter_y = [2, 4, 6, 8, 10]
    pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
    pie_sizes = [40, 25, 20, 15]

    # Create a 2x2 subplot layout
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # First subplot: Line plot
    axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
    axs[0, 0].set_title("Line Plot")
    axs[0, 0].set_xlabel("X Values")
    axs[0, 0].set_ylabel("Y Values")
    axs[0, 0].grid(True, linestyle="--", alpha=0.6)

    # Second subplot: Bar plot
    axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axs[0, 1].set_title("Bar Plot")
    axs[0, 1].set_xlabel("Categories")
    axs[0, 1].set_ylabel("Values")
    axs[0, 1].grid(axis='y', linestyle="--", alpha=0.6)

    # Third subplot: Scatter plot
    axs[1, 0].scatter(x, scatter_y, color='blue', marker='o')
    axs[1, 0].set_title("Scatter Plot")
    axs[1, 0].set_xlabel("X Values")
    axs[1, 0].set_ylabel("Y Values")
    axs[1, 0].grid(True, linestyle="--", alpha=0.6)

    # Fourth subplot: Pie chart
    axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'], startangle=140)
    axs[1, 1].set_title("Pie Chart")

    # Adjust layout for better spacing
    plt.tight_layout()

    # Show the plots
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
